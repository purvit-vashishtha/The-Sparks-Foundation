## Data Visualization

```python
# Countplot of attacks in recent years
plt.figure(figsize=(18,6))
sns.countplot(df['iyear'])
plt.xlabel("Attacks frequency")
plt.xticks(rotation=90)
plt.ylabel("Number of Attacks")
plt.title("No of Attacks in year (1970-2017)")
```

```python
plt.figure(figsize=(18,8))
sns.countplot(x=df["iyear"], hue="success", data = df)
plt.xlabel("Years")
plt.xticks(rotation=90)
plt.ylabel("Frequency of Attacks")
```

```python
# bar plot to show countries with most no. of terrorism
plt.figure(figsize=(18,8))
sns.barplot(x=df['country_txt'].value_counts()[:20].index, y=df['country'].value_counts()[:20])
plt.xlabel("Countries")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.title("Countries facing Most Terrorism")
```

```python
# bar plot to show Regions with most terrorism
plt.figure(figsize=(18,8))
sns.countplot(x= df["region_txt"], hue="success", data=df)
plt.xlabel("Regions", fontsize=20)
plt.xticks(rotation=90)
plt.ylabel("Terrorism Count", fontsize=20)
plt.title("Regions facing Most Terrorism", fontsize=20)
```

```python
# barplot of waepons used
plt.figure(figsize=(14,7))
sns.barplot(x= df["weaptype1_txt"].value_counts()[:50].index,y = df["weaptype1_txt"].value_counts()[:50].values)
plt.xlabel("Weapons used", fontsize=20)
plt.xticks(rotation=90)
plt.ylabel("Frequency of a Particluar weapon used", fontsize=12)
plt.title("Types of Weapons", fontsize=20)
```

```python
# barplot of attacktypes
plt.figure(figsize=(12,8))
sns.barplot(x=df['attacktype1_txt'].value_counts()[:30].index,y=df['attacktype1_txt'].value_counts()[:30].values)
plt.xlabel("Attacks", fontsize=20)
plt.xticks(rotation=90)
plt.ylabel("Count of a particular Attack", fontsize=20)
plt.title("Types of Attacks", fontsize=20)
```
